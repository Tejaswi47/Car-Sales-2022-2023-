```python
car_model_sold = {}
for i in cars_model:
    car_model_sold[i] = data[data['Car Model'] == i].shape[0]
print(car_model_sold)
```

```python
maximum_car_sale = data.shape[0]
```

```python
car_sold_percentage = {}
for i in car_model_sold:
    car_sold_percentage[i] = (car_model_sold[i]/ maximum_car_sale) * 100 
print(car_sold_percentage)
```

```python
colors = ['#ff8080', '#66ffc2', '#d9ff66', '#ffb366', '#66ccff']
explode = [0.01, 0.01, 0.01, 0.1, 0.01]
```

```python
plt.style.use('fivethirtyeight')

plt.pie(car_sold_percentage.values(),labels = cars_model ,colors= colors, shadow= True, wedgeprops= {'edgecolor' : 'black'}, explode= explode, autopct='%1.3f%%')

plt.title('Cars Sold percentage in 2023-2024')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.title('Number of Cars Sold by Model')
plt.xlabel('Car Model')
plt.yscale('log')
plt.ylabel('Number of Cars Sold')
plt.bar(cars_model, car_model_sold.values())
```
